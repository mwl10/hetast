```python
import PyROA
import os
```

```python
base_dir = '/Users/mattlowery/Desktop/code/astro/javelin/my_examples/dat/AGN_1H2106-099/epochs'
epoch_dirs = [os.path.join(base_dir, epoch, '') for epoch in os.listdir(base_dir)]
epoch_dirs.sort()
print(epoch_dirs, len(epoch_dirs))

filters = ["U", "G", "R", "I", "Z"] 
objName = "AGN_1H2106-099"
psi_types = ["LogGaussian","LogGaussian", "LogGaussian", "LogGaussian"]

            # rms      # mean       # lag        # width of window  # extra error 
priors = [[0.0, 20.0], [0.0, 20.0], [-50.0, 50.0], [0.01, 10.0], [0.0, 10.0]]
#priors = [[0.0, 50.0],[0.0, 1000.0], [-50.0, 50.0], [0.01, 10.0], [0.0, 50.0]]

# epoch 1
fit = PyROA.Fit(epoch_dirs[2], objName, filters, priors, delay_dist=True, add_var=True, psi_types=psi_types)
PyROA.Plot(fit)
```
